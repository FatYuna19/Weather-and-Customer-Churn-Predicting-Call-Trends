```python
# Prepare your feature set and target variable
X = merged_data.drop(columns=['churn'])

# Select only numeric columns
X = X.select_dtypes(include=[np.number])
y = merged_data['churn']

# Add a constant to the model
X_with_const = sm.add_constant(X)

# Fit the model
model = sm.OLS(y, X_with_const).fit()

# Get the p-values
p_values = model.pvalues

# Set the p-value threshold
threshold = 0.05

# Identify features to drop (excluding 'const')
to_drop = p_values[p_values > threshold].index.tolist()
if 'const' in to_drop:
    to_drop.remove('const')  # Remove constant if it's there

# Drop features from the numeric feature set
cleaned_X = X.drop(columns=to_drop)

# Combine cleaned features with the target variable
cleaned_data = pd.concat([cleaned_X, y], axis=1)

# Display dropped features and new shape
print("Dropped features with p-values above threshold:")
print(to_drop)
print("\nNew shape of the dataset:", cleaned_data.shape)
```

```python
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan

# Define your dependent and independent variables
X = cleaned_data.drop('churn', axis=1)  # Independent variables
y = cleaned_data['churn']  # Dependent variable

# Add a constant to the independent variables
X = sm.add_constant(X)

# Fit the model
model = sm.OLS(y, X).fit()

# Perform the Breusch-Pagan test
bp_test = het_breuschpagan(model.resid, model.model.exog)
bp_test_results = pd.Series(bp_test, index=['LM Statistic', 'LM p-value', 'F Statistic', 'F p-value'])

print("Breusch-Pagan Test Results:")
print(bp_test_results)
```

```python
# Fit the model with robust standard errors
model_robust = sm.OLS(y, X).fit(cov_type='HC3')

# Display the summary
print(model_robust.summary())
```

```python
from scipy.stats import shapiro

# Get the residuals from the model
residuals = model.resid

# Perform the Shapiro-Wilk test
shapiro_stat, shapiro_p = shapiro(residuals)

print(f"Shapiro-Wilk Test Statistic: {shapiro_stat}, p-value: {shapiro_p}")

# Interpretation
if shapiro_p > 0.05:
    print("The residuals are normally distributed (fail to reject H0).")
else:
    print("The residuals are not normally distributed (reject H0).")
```

```python
from sklearn.feature_selection import RFE
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectFromModel

# Define features (X) and target (y)
X = cleaned_data.drop(columns=['churn', 'churn_log', 'churn_sqrt', 'churn_boxcox'])  # Remove target variables
y = cleaned_data['churn']

# RFE with SVC (linear kernel)
svc_model = SVC(kernel='linear')
rfe_selector = RFE(estimator=svc_model, n_features_to_select=5)  # Choose the number of features
rfe_selector.fit(X, y)

# Display selected features by RFE
rfe_selected_features = X.columns[rfe_selector.support_]
print("RFE Selected Features:", rfe_selected_features)

# L1 Regularization (Lasso) with Logistic Regression
logistic_model = LogisticRegression(penalty='l1', solver='liblinear')
lasso_selector = SelectFromModel(logistic_model)
lasso_selector.fit(X, y)

# Display selected features by Lasso
lasso_selected_features = X.columns[lasso_selector.get_support()]
print("Lasso Selected Features:", lasso_selected_features)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Create a DataFrame for feature comparison
feature_comparison = pd.DataFrame({
    'Features': ['international_plan', 'humidity', 'service_calls_wind', 'churn_wind', 'combined_service_calls_humidity'],
    'RFE_Selected': [1 if feature in rfe_selected_features else 0 for feature in ['international_plan', 'humidity', 'service_calls_wind', 'churn_wind', 'combined_service_calls_humidity']],
    'Lasso_Selected': [1 if feature in lasso_selected_features else 0 for feature in ['international_plan', 'humidity', 'service_calls_wind', 'churn_wind', 'combined_service_calls_humidity']]
})

# Plot the results
feature_comparison.plot(x='Features', kind='bar', figsize=(10, 6), width=0.8)
plt.title('Feature Selection Comparison: RFE vs Lasso')
plt.ylabel('Selected (1 = Yes, 0 = No)')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression, RidgeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Use the RFE-selected features
X_rfe = cleaned_data[rfe_selected_features]
y = cleaned_data['churn']

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_rfe, y, test_size=0.2, random_state=42)

# Define a list of all applicable models to test
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Random Forest': RandomForestClassifier(),
    'SVC': SVC(kernel='linear', probability=True),
    'Decision Tree': DecisionTreeClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Naive Bayes': GaussianNB(),
    'Ridge Classifier': RidgeClassifier()
}

# Train and evaluate each model
for model_name, model in models.items():
    print(f"\nTraining and evaluating {model_name}...\n")
    
    # Train the model
    model.fit(X_train, y_train)
    
    # Predict on the test set
    y_pred = model.predict(X_test)
    
    # Display classification report
    print(f"Classification Report for {model_name}:")
    print(classification_report(y_test, y_pred))
    
    # Confusion matrix
    conf_matrix = confusion_matrix(y_test, y_pred)
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix for {model_name}')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grids for KNN and Ridge Classifier
param_grid_knn = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski']
}

param_grid_ridge = {
    'alpha': [0.1, 1, 10],
    'solver': ['auto', 'sag', 'lsqr']
}

# Initialize the models
knn = KNeighborsClassifier()
ridge = RidgeClassifier()

# Perform grid search for KNN
grid_search_knn = GridSearchCV(knn, param_grid_knn, cv=5, scoring='accuracy')
grid_search_knn.fit(X_train, y_train)

# Perform grid search for Ridge Classifier
grid_search_ridge = GridSearchCV(ridge, param_grid_ridge, cv=5, scoring='accuracy')
grid_search_ridge.fit(X_train, y_train)

# Get the best parameters and scores
print(f"Best Parameters for KNN: {grid_search_knn.best_params_}")
print(f"Best CV Accuracy for KNN: {grid_search_knn.best_score_}")

print(f"Best Parameters for Ridge Classifier: {grid_search_ridge.best_params_}")
print(f"Best CV Accuracy for Ridge Classifier: {grid_search_ridge.best_score_}")

# Evaluate on the test set
best_knn = grid_search_knn.best_estimator_
best_ridge = grid_search_ridge.best_estimator_

# Predictions
y_pred_knn = best_knn.predict(X_test)
y_pred_ridge = best_ridge.predict(X_test)

# Classification reports and confusion matrices
print("Classification Report for KNN:")
print(classification_report(y_test, y_pred_knn))
print("Confusion Matrix for KNN:")
sns.heatmap(confusion_matrix(y_test, y_pred_knn), annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix for KNN')
plt.show()

print("Classification Report for Ridge Classifier:")
print(classification_report(y_test, y_pred_ridge))
print("Confusion Matrix for Ridge Classifier:")
sns.heatmap(confusion_matrix(y_test, y_pred_ridge), annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix for Ridge Classifier')
plt.show()
```
